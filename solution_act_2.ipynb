# visualización data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly
```

```python
fig, ax = plt.subplots(1, 1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title("Gráfico 1")
sns.barplot(data=act_2,
            x='ROOMS',
            y='PRICE',
            hue='YEAR',
            palette='viridis',
            ax=ax )
```

En este gráfico se evidencia como los precios de las casas aumentan según la cantidad de las habitaciones para los años 1978 y 1981
